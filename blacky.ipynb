```python
import cv2
import numpy as np
from PIL import Image
from glob import glob
import matplotlib.pyplot as plt
```

```python
targetPath = "Dataset/Normal/"
savePath = "Dataset/Normal_label_black/"
```

```python
target = glob(targetPath + '*.png')
len(target)
```

```python
for item in target :
#     img = Image.open(item)
#     img = np.array(img)
#     t_img = cv2.imread(item,0)
    
#     print(img)
#     print(t_img)
    
    imgName = item[item.find("/T1") + 1 : -4]
        
    black = np.zeros((256, 256, 3))
    
    black[:,:,:] = 0
    
    img = Image.fromarray(black.astype(np.uint8))
    img.save(savePath + imgName + '.png' , 'PNG')
    
```

## Replace

```python
import os
import shutil
```

```python
targetPath = "Dataset/Normal_label_black/"
savePath = "resource/dataset_ev/label/"
```

```python
labels = os.listdir(targetPath)
print(len(labels))
```

```python
r_labels = os.listdir(savePath)
print(len(r_labels))
```

```python
count = 0
for label in labels :
    if label in r_labels :
        os.remove(savePath + label)
        shutil.copy(targetPath + label , savePath + label)
        print(str(count) + " " + label)
        count += 1 
```
